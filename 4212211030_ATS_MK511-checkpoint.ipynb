### Library and Load Images Data

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import LeaveOneOut, cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from skimage.feature import hog
from mlxtend.data import loadlocal_mnist
from mlxtend.plotting import plot_confusion_matrix
```

```python
train_images, train_labels = loadlocal_mnist(images_path='EMNIST/emnist_source_files/emnist-bymerge-train-images-idx3-ubyte',
                                             labels_path='EMNIST/emnist_source_files/emnist-bymerge-train-labels-idx1-ubyte')
```

```python
test_images, test_labels = loadlocal_mnist(images_path='EMNIST/emnist_source_files/emnist-bymerge-test-images-idx3-ubyte',
                                           labels_path='EMNIST/emnist_source_files/emnist-bymerge-test-labels-idx1-ubyte')
```

### Train Data

```python
train_images_sampled = train_images[:2000]
train_labels_sampled = train_labels[:2000]
```

```python
n_samples = train_images_sampled.shape[0]
```

```python
feature, hog_img = hog(
    train_images[0].reshape(28, 28),
    orientations=9,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2),
    visualize=True,
    block_norm='L2'
)
hog_feature_length = feature.shape[0]
```

```python
feature.shape
```

```python
X_train = np.zeros((n_samples, hog_feature_length))
y_train = np.zeros(n_samples, dtype=train_labels_sampled.dtype)
```

```python
for i in range(n_samples):
    X_train[i], _ = hog(
        train_images_sampled[i].reshape(28, 28),
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        visualize=True,
        block_norm='L2'
    )
    y_train[i] = train_labels_sampled[i]
```

```python
scaler = StandardScaler()
```

```python
hog_features_train_scaled = scaler.fit_transform(X_train)
```

```python
clf = svm.SVC(kernel='rbf', C=1, gamma='scale')
```

```python
loo = LeaveOneOut()
y_pred = cross_val_predict(clf, hog_features_train_scaled, y_train, cv=loo)
loo_confusion_matrix = confusion_matrix(y_train, y_pred)
loo_accuracy = accuracy_score(y_train, y_pred)
loo_precision = precision_score(y_train, y_pred, average='weighted', zero_division=1)
loo_recall = recall_score(y_train, y_pred, average='weighted', zero_division=1)
loo_f1 = f1_score(y_train, y_pred, average='weighted', zero_division=1)
```

```python
print("\nLOOCV Results (Training Data):")
print("Confusion Matrix:")
print(loo_confusion_matrix)
print("Accuracy:", loo_accuracy)
print("Precision:", loo_precision)
print("Recall:", loo_recall)
print("F1 Score:", loo_f1)
```

### Test Data

```python
test_images_sampled = test_images[:2000]
test_labels_sampled = test_labels[:2000]
```

```python
n_samples = test_images_sampled.shape[0]
```

```python
X_test = np.zeros((n_samples, hog_feature_length)) 
y_test = np.zeros(n_samples, dtype=test_labels_sampled.dtype)
```

```python
for i in range(n_samples):
    X_test[i], _ = hog(
        test_images_sampled[i].reshape(28, 28),
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        visualize=True,
        block_norm='L2'
    )
    y_test[i] = test_labels_sampled[i]
```

```python
hog_features_test_scaled = scaler.fit_transform(X_test)
```

```python
loo = LeaveOneOut()
y_pred = cross_val_predict(clf, hog_features_test_scaled, y_test, cv=loo)
loo_confusion_matrix = confusion_matrix(y_test, y_pred)
loo_accuracy = accuracy_score(y_test, y_pred)
loo_precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
loo_recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)
loo_f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)
```

```python
def plot_confusion_mat(cm):
  fig, ax = plt.subplots(figsize=(10, 10))
  plot_confusion_matrix(conf_mat=cm, ax=ax, cmap=plt.cm.Blues)
  plt.title('Confusion Matrix')
  plt.xlabel('Predicted Label')
  plt.ylabel('True Label')
  plt.show()
```

```python
plot_confusion_matrix(loo_confusion_matrix)
```

```python
def plot_images_with_hog(original_images, hog_images, title=""):
    fig, axes = plt.subplots(2, 10, figsize=(15, 5))
    fig.suptitle(title)

    for i in range(min(len(original_images), 10)):
        axes[0, i].imshow(original_images[i].reshape((28, 28)), cmap='gray')
        axes[0, i].axis('off')

    for i in range(min(len(hog_images), 10)):
        axes[1, i].imshow(hog_images[i], cmap='gray')
        axes[1, i].axis('off')

    plt.show()
```

```python
def extract_hog_images(images):
  hog_images = []
  for image in images:
    _,  hog_img = hog(image.reshape((28, 28)),
                      orientations=9,
                      pixels_per_cell=(8, 8),
                      cells_per_block=(2, 2),
                      visualize=True,
                      block_norm='L2')
    hog_images.append(hog_img)
  return np.array(hog_images)
```

```python
hog_images_test = extract_hog_images(test_images_sampled)
```

```python
plot_images_with_hog(test_images_sampled, hog_images_test, title='Original Images vs Extracted HOG Features')
```
